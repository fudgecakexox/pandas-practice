```python
import numpy as np
import pandas as pd
```

```python
# Create a Pandas Series from a list.

a = pd.Series([1, 2, 3, 4, 5])
a
```

```python
# Create a Series with a custom index.

marks = [98, 90, 97, 100, 95]
subjects= ['Science', 'Math', 'Hindi', 'English', 'Computer Science']

marks_series = pd.Series(marks, index = subjects)

marks_series
```

```python
# Create a DataFrame from:

# Dictionary of lists
a1 = {
    "name": ['Ananya', 'Aditi', 'Shreyansh'],
    "age": [23, 21, 14],
    "gender": ['female', 'female', 'male']
}

info_df = pd.DataFrame(a1)

info_df

# NumPy array
b = np.arange(12).reshape(4, 3)
b_df = pd.DataFrame(b)

b_df
```

```python
# Display head(), tail(), info(), describe().

print(info_df.head(2))
print(info_df.tail(2))

print(b_df.info())
print(b_df.describe())
print(info_df.describe())
```
